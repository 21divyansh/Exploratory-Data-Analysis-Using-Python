```python
import pandas as pd
emp_df=pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
emp_df
```

```python
dep_data=emp_df.groupby('Department')['Annual Salary ($)'].sum()
dep_data
```

```python
import matplotlib.pyplot as plt
```

```python
dep_data.plot(x='Department',kind='bar',stacked=False,title='Grouped bar chart',
xlabel='Department',ylabel='Total Annual salary')
```

```python
# Interpretation: The IT department has the employees with highest annual salary i.e, 1,28,2900/-
```
