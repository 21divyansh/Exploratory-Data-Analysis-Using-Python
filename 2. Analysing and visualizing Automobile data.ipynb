## Task 1: Identify the top 10 cars manufacturers based on the city mileage and display the result using a horizantal bar graph with the manufaturers on the yaxis and mileage on the x axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
car_data=pd.read_excel('DS1_C4_S5_Car_Data_Challenge.xlsx')
car_data
```

```python
car_data1=car_data.groupby('Make')['City_Mileage_km_litre'].max().sort_values(ascending=False).head(10)
car_data1
```

```python
car_data1.plot(x='Make',kind='barh',stacked=False,title='Horizantal Bar Graph',xlabel='City_Mileage',ylabel='Manufacturers')
```

#### Interpretation: The above visual show that the 'City Mileage' in 'Mahindra' is extremely high that interprets that it has outliers present in it.
