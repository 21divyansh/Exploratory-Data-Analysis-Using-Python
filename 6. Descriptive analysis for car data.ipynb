# Task 2: Preprocessing the data in the following columns as specified.

```python
df['Kerb_Weight']=df['Kerb_Weight'].str.replace(r"kg",'')
df
# Kerb_Weight
```

```python
df['Kerb_Weight']=df['Kerb_Weight'].astype(float)
df['Kerb_Weight'].dtypes
```

```python
df['Kerb_Weight']=df.loc[:,['Kerb_Weight']]*2.2
df
```

```python
df.loc[:,['Kerb_Weight']]
```

```python
df["Power"] = df["Power"].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*", '', regex=True).astype(float)
df["Power"]
```

```python
df['Torque'] = df['Torque'].str.replace(r"-",'', regex=True)
df["Torque"] = df["Torque"].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*", '', regex=True).astype(float)
df["Torque"]
```

```python
mean = []
median = []
sd = []
cv = []

manufacturers = df1['Make'].unique()

for manufacture in manufacturers:
    print(f"\nManufacturer: {manufacture}")
    for i in df1.columns:
        if i != 'Make':
            values = df1.loc[df1['Make'] == manufacture, i]
            mean1 = values.mean()
            median1 = values.median()
            sd1 = values.std()
            cv1 = mean1 / sd1 * 100

            mean.append(mean1)
            median.append(median1)
            sd.append(sd1)
            cv.append(cv1)
            print(f"\nVariable: {i}")
            print(f"Mean: {mean1}")
            print(f"Median: {median1}")
            print(f"Standard Deviation: {sd1}")
            print(f"Coefficient of Variation: {cv1}")
```
