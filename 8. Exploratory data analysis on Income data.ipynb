# Task 2: Create two different lists of categorical columns and numerical columns by defining a function to identify the data type. 

```python
def sep_df(df):
    categorical=[]
    numerical=[]
    for i in df.columns:
        if df[i].dtypes == 'O':
            categorical.append(i)
        else:
            numerical.append(i)
    return categorical,numerical
categorical, numerical = sep_df(df) 
```

# Task 3: Define a function to find the unique values and mode value from each categorical column 

```python
def cat_info(i):
    unique_values = df[i].unique()
    
    if ' ?' in unique_values:
        df[i].replace(' ?', df[i].mode()[0], inplace=True)
    
    print(f'The total number of unique values in {i} is: {df[i].nunique()}')
    print(f'The mode value of {i} is: {df[i].mode()[0]}')
```

# Task 5 Perform descriptive analysis of each numerical column by defining a function to print the following information.

```python
def ct_data(i):
    print(f'The mean of {i} is : {df[i].mean()}')
    print(f'The median of {i} is : {df[i].median()}')
    print(f'The standard deviation of {i} is : {df[i].std()}')
    print(f'The number of missing values in {i} is : {df[i].isnull().sum()}')
    print(f'The skewness of {i} is : {df[i].skew()}')
```

# Task 6: Perform descriptive analysis of the numerical columns with the help of univariate graphs.

```python
def ct_data(i):
    mean = df[i].mean()
    std = df[i].std()
    lower_bound = mean - 2 * std
    upper_bound = mean + 2 * std

    df[i] = np.where((df[i] < lower_bound) | (df[i] > upper_bound), mean, df[i])

    print(f'The mean of {i} is : {df[i].mean()}')
    print(f'The median of {i} is : {df[i].median()}')
    print(f'The standard deviation of {i} is : {df[i].std()}')
    print(f'The number of missing values in {i} is : {df[i].isnull().sum()}')
    print(f'The skewness of {i} is : {df[i].skew()}')
```

```python
ct_data('Age')
```

```python
fig, ax = plt.subplots(1, 2, figsize= (20, 7))                
sns.histplot(x = df["Age"], ax =ax[1], color = "m") 
sns.boxplot(x = df["Age"], ax = ax[0], color = "lightgreen") 
plt.show()
```

## Interpretation: After the treatment of outliers the above visual shows there are no outliers present in the data.

```python
ct_data('Education_Num')
```

```python
fig, ax = plt.subplots(1, 2, figsize= (20, 7))                
sns.histplot(x = df['Education_Num'], ax =ax[1], color = "orange") 
sns.boxplot(x = df['Education_Num'], ax = ax[0], color = "lightgreen") 
plt.show()
```

## Interpretation: After the treatment of outliers the above visual shows there are no outliers present in the data but the it is slightly right skewed.
