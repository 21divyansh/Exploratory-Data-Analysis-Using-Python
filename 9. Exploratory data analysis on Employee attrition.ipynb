# Task 1: Perform bivariate and multivariate analysis on categorical columns with the help of graphs/charts, as mentioned.

```python
import matplotlib.pyplot as plt
from tabulate import tabulate
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('DS1_C5_S2_EmployeeAttrition_Data_Practice.csv')
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
def sep_df(df):
    categorical=[]
    numerical=[]
    for j in df.columns:
        if df[j].nunique() <=6:
            categorical.append(j)
        else:
            numerical.append(j)
    return categorical,numerical
categorical,numerical = sep_df(df)

print(tabulate({'Categorical':categorical,'Numerical':numerical},headers=['Categorical','Numerical']))
```

```python
# Business category wise attrition of employees.
fig,ax=plt.subplots(figsize=(10,5))
ax.set_title('Business category wise attrition of employees')
sns.countplot(x='BusinessTravel',hue='Attrition',data=df)
plt.show()
```

## Interpretation: The above visual shows that there are most of the employees who travel rarely and also attrited as compare to the business travellers and non travellers.

```python
# Education field wise attrition of employees.
fig,ax=plt.subplots(figsize=(10,5))
ax.set_title('Education field wise attrition of employees')
sns.countplot(x='EducationField',hue='Attrition',data=df)
plt.show()
```

# Task 2: Perform the bivariate analysis between categorical and numerical variables using an appropriate graphs.

```python
df.isnull().sum()
```

```python
# Age wise attrition of employees.
fig, ax = plt.subplots(1, 2, figsize = (15, 7))
sns.histplot(x = df["Age"], hue = df["Attrition"], ax = ax[0])
sns.boxplot(y = df["Age"], x = df["Attrition"], ax = ax[1])
plt.show()
```

## Interpretation: The above visual shows that most are the employees are between 30 to 40 years of age but the employees between 25 to 40 are attritting more. Theres also right skewness in the data and two outliers.
